```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def some(array):
    for i in array:
        if i != True:
            return False
    return True
f'Test False: {some(np.array([True, True, False, True]))}, Test True: {some(np.array([True, True, True, True]))}'
```

### Sigmoid

```python
def sigmoid(lamb, x):
    return 1 / (1 + np.exp(-lamb * x))
```

```python
x = np.linspace(-10, 10, 20)
y = sigmoid(0.5, x)
plt.plot(x, y)

y = sigmoid(1, x)
plt.plot(x, y)

y = sigmoid(5, x)
plt.plot(x, y)

plt.legend(('0.5', '1.0', '5.0'))
```

### Hyperbolic Tangent

```python
def cosh(x):
    return (np.exp(x) + np.exp(-x)) / 2


def sinh(x):
    return (np.exp(x) - np.exp(-x)) / 2


def tanh(x):
    return sinh(x) / cosh(x)
```

```python
x = np.linspace(-10, 10, 20)
# y = sinh(x)
# plt.plot(x, y)

# x = cosh(x)
# plt.plot(x, y)

y = tanh(0.5 * x)
plt.plot(x, y)

y = tanh(1 * x)
plt.plot(x, y)

y = tanh(5 * x)
plt.plot(x, y)


plt.legend(('0.5', '1.0', '5.0'))
```

### Rectified Linear Unit (ReLU)

```python
def relu(x):
    for (idx, value) in enumerate(np.nditer(x)):
        x[idx] = max(0.0, value)
    return x
```

```python
x = np.linspace(-10, 10, 20)
y = relu(0.5 * x)
plt.plot(x, y)

y = relu(1 * x)
plt.plot(x, y)

y = relu(5 * x)
plt.plot(x, y)

plt.legend(('0.5', '1.0', '5.0'))
```

### Sigmoid + ReLU (Interesting)

```python
x = np.linspace(-10, 10, 20)
y = sigmoid(0.5, relu(0.5 * x))
plt.plot(x, y)

y = sigmoid(1, relu(1 * x))
plt.plot(x, y)

y = sigmoid(5, relu(5 * x))
plt.plot(x, y)

plt.legend(('0.5', '1.0', '5.0'))
```

```python
x = np.linspace(-10, 10, 20)
y = relu(sigmoid(0.5, x))
plt.plot(x, y)

y = relu(sigmoid(1, x))
plt.plot(x, y)

y = relu(sigmoid(5, x))
plt.plot(x, y)

plt.legend(('0.5', '1.0', '5.0'))
```

### Sigmoid * ReLU + Hyperbolic Tangent (Interesting)

```python
x = np.linspace(-10, 10, 20)
y = sigmoid(0.5, x) * relu(tanh(0.5 * x)) 
plt.plot(x, y)

y = sigmoid(1, x) * relu(tanh(1 * x))
plt.plot(x, y)

y = sigmoid(5, x) * relu(tanh(5 * x))
plt.plot(x, y)

plt.legend(('0.5', '1.0', '5.0'))
```

### ReLU + Sigmoid is same result than Sigmoid ? Yes

```python
x = np.linspace(-10, 10, 20)
yReluSigmoid = relu(sigmoid(0.5, x))
ySigmoid = sigmoid(0.5, x)
yResult0_5 = yReluSigmoid == ySigmoid

yReluSigmoid = relu(sigmoid(1, x))
ySigmoid = sigmoid(1, x)
yResult1_0 = yReluSigmoid == ySigmoid

yReluSigmoid = relu(sigmoid(5, x))
ySigmoid = sigmoid(5, x)
yResult5_0 = yReluSigmoid == ySigmoid

f'0.5: {some(yResult0_5)}, 1.0: {some(yResult1_0)}, 5.0: {some(yResult5_0)}'
```

### Hyperbolic Tangent + ReLU is same result than ReLU + Hyperbolic Tangent ? Yes

```python
x = np.linspace(-10, 10, 20)
yTanhRelu = tanh(relu(0.5 * x))
yReluTanh = relu(tanh(0.5 * x))
yResult0_5 = yTanhRelu == yReluTanh


yTanhRelu = tanh(relu(1 * x))
yReluTanh = relu(tanh(1 * x))
yResult1_0 = yTanhRelu == yReluTanh


yTanhRelu = tanh(relu(5 * x))
yReluTanh = relu(tanh(5 * x))
yResult5_0 =yTanhRelu == yReluTanh

f'0.5: {some(yResult0_5)}, 1.0: {some(yResult1_0)}, 5.0: {some(yResult5_0)}'
```

### Hyperbolic Tangent + Sigmoid is same than Sigmoid ? No

```python
x = np.linspace(-10, 10, 20)
yTanhSigmoid = tanh(sigmoid(0.5, x))
ySigmoid = sigmoid(0.5, x)
yResult0_5 = yTanhSigmoid == ySigmoid

yTanhSigmoid = tanh(sigmoid(1, x))
ySigmoid = sigmoid(1, x)
yResult1_0 = yTanhSigmoid == ySigmoid

yTanhSigmoid = tanh(sigmoid(5, x))
ySigmoid = sigmoid(5, x)
yResult5_0 = yTanhSigmoid == ySigmoid

f'0.5: {some(yResult0_5)}, 1.0: {some(yResult1_0)}, 5.0: {some(yResult5_0)}'
```

### Sigmoid * ReLU + Hyperbolic Tangent is same than Hyperbolic Tangent + ReLU ? No

```python
x = np.linspace(-10, 10, 20)
ySigmoidTanhRelu = sigmoid(0.5, x) * relu(tanh(0.5 * x)) 
yTahnRelu = tanh(relu(0.5 * x))
yResult0_5 = ySigmoidTanhRelu == yTahnRelu

ySigmoidTanhRelu = sigmoid(1, x) * relu(tanh(1 * x)) 
yTahnRelu = tanh(relu(1 * x))
yResult1_0 = ySigmoidTanhRelu == yTahnRelu

ySigmoidTanhRelu = sigmoid(5, x) * relu(tanh(0.5 * x)) 
yTahnRelu = tanh(relu(5 * x))
yResult5_0 = ySigmoidTanhRelu == yTahnRelu

f'0.5: {some(yResult0_5)}, 1.0: {some(yResult1_0)}, 5.0: {some(yResult5_0)}'
```
